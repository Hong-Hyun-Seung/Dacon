#### 대회 평가지표
월간 데이콘 2회 대회의 평가지표는 Logloss 입니다. 강의 시간에 배운 평가 지표로는 정확도, 정밀도, 재현율, F1_Score 등이 있었는데 이러한 평가 지표들은 라벨 값을 이용해 계산하게 됩니다. <br>
하지만 Logloss는 라벨 값이 아닌 모델이 해당 클래스를 예측한 확률을 이용해 평가하는 지표입니다. Logloss는 Cross Entropy라고도 불리며 주로 분류 모델의 손실 함수로 사용됩니다. <br>
따라서 결과를 만드실 때, predict() 함수가 아닌 predict_proba() 라는 함수를 사용해야합니다. 제출 파일을 생성하는 코드는 남겨드리니 확률 값만 넣어주시면 됩니다.<br>
남겨드린 참고 링크를 확인해보시고, 이해가 가지 않는 부분은 질문 부탁드리겠습니다.

```python
x_train, x_valid, y_train, y_valid = train_test_split(train_data, label, 
                                                      test_size=0.3,
                                                      random_state=2021,
                                                      shuffle=True,
                                                      stratify=label)
```

#### 2. 분류 모델 사용해보기
1) Logistic Regression <br>
2) Support Vector Machine <br>
3) Random Forest(Decision Tree) <br>
4) XGBoost <br>

- 다섯가지 분류 모델의 성능을 비교해보세요. <br>
- 해당 대회의 평가지표는 Logloss 이므로 "Scikit-learn Logloss"라고 검색하여 해당 평가 지표를 사용해 Validation 성능을 비교합니다.<br> 그 중 가장 좋은 성능의 모델을 선택해 Test 데이터의 결과를 예측합니다. Log loss는 손실 값이니 낮을 수록 좋은 모델입니다.
- 모델 파라미터는 지난 수업에서 제가 추천해드렸던 파라미터의 값을 실제로 변경해보고, 모델 튜닝도 시도해보시는 것을 추천드립니다. <br> 강의 시간에도 답변드렸지만, 실제로 파라미터 튜닝은 컴퓨터가 최적화하도록 하는 것이 효율적이니 한번 해보는 정도로만 해보셔도 괜찮습니다. (random_state 값은 고정시키는 것을 강력히 추천드립니다.)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import log_loss

models = [['Logistic Regression', LogisticRegression(random_state=2020,
                                                     n_jobs=-1)], 
#           ['Support Vector Machine', SVC(probability=True)], 
          ['RandomForest', RandomForestClassifier(random_state=2020,
                                                  n_jobs=-1)], 
          ['XGBoost', XGBClassifier(random_state=2020,
                                    n_jobs=4)], 
          ['LightGBM', LGBMClassifier(random_state=2020,
                                      n_jobs=4)]]

best_loss = np.inf

for i, (name, model) in enumerate(models):
    model.fit(x_train, y_train)
    y_pred_proba = model.predict_proba(x_valid)
    current_loss = log_loss(y_valid, y_pred_proba)
    
    if current_loss < best_loss:
        best_loss  = current_loss
        best_model = [i, name]
    
    print(f'{name} Model Validation Logloss: {current_loss}')

print(f'Best Model: {best_model[-1]}, Logloss: {best_loss}')
```

```python
pred = models[best_model[0]][-1].predict_proba(x_test)
```
