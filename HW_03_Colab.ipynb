##### 중요!
- 이번 데이터는 라벨의 개수가 4개이므로 각 모델을 테스트할 때 ['layer_1','layer_2','layer_3','layer_4'] 각 라벨에 대한 4개 모델을 생성하셔야합니다. XGBoost와 LightGBM은 다중 출력 회귀 모델을 지원하지만, 이번 과제에서는 4개 모델을 생성하시길 권해드립니다. 
- 모델을 많이 만들기 때문에 모델 정의 시 병렬 처리를 할 수 있도록 n_jobs 옵션이 있는 모델의 경우 반드시 설정하시기 바랍니다.
- 해당 대회의 평가지표는 MAE 이므로 "Scikit-learn MAE"라고 검색하여 해당 평가 지표를 사용해 Validation 성능을 비교합니다.<br> 그 중 가장 좋은 성능의 모델을 선택해 Test 데이터의 결과를 예측합니다. MAE는 손실 값이니 낮을 수록 좋은 모델입니다.
    - 각 모델에 대해 4개 라벨의 MAE를 평균내어 가장 낮은 모델을 선택하고, 결과를 제출해보세요.
    - 가능 하다면 각 라벨마다 성능이 제일 좋은 모델을 선택해보세요. ex) 'layer_1': xgboost, 'layer_2': lightgbm, 'layer_3': randomforest, 'layer_4':xgboost

```python
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error

models = [
          # ['Linear Regression', LinearRegression(n_jobs=-1)], 
          # ['MultiLayerPerceptron', MLPRegressor(random_state=2021)],
          # ['RandomForest', RandomForestRegressor(random_state=2021,
          #                                         n_jobs=-1)], 
          ['XGBoost', XGBRegressor(random_state=2021,
                                   tree_method='gpu_hist',
                                    n_jobs=7)], 
          ['LightGBM', LGBMRegressor(random_state=2021,
                                   tree_method='gpu_hist',
                                      n_jobs=7)]]


best_loss = np.inf
best_pred = None
for i, (name, model) in enumerate(models):
    current_loss = list()
    current_pred = list()
    for c_label in label_columns:
        model.fit(x_train, y_train[c_label])
        y_pred = model.predict(x_valid)
        current_loss.append(mean_absolute_error(y_valid[c_label], y_pred))
        current_pred.append(model.predict(x_test))
        
    if sum(current_loss) < best_loss:
        best_loss  = sum(current_loss)
        best_model = [i, name]
        best_pred  = current_pred
    
    print(f'{name} Model Validation Mean Absolute Error: {sum(current_loss)}')

print(f'Best Model: {best_model[-1]}, Mean Absolute Error: {best_loss}')
```
